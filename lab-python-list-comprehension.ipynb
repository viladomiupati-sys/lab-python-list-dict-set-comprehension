````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
    return inventory
````

````python
def update_inventory(customer_orders, inventory):
    update_inventory = {product : (quantity -1 if product in customer_orders else quantity)
                        for product, quantity in customer_orders.items()}
    return update_inventory
````

````python
def print_updated_inventory(inventory):
    print_updated_inventory = {{product}: {quantity} for product, quantity in inventory.items()}
    return print_updated_inventory
````

````python
def get_customer_orders():
    user_order = int(input("Enter the number of customer orders: "))

    order = [input("Enter the name of a product that a customer wants to order:  ")
            for i in range(user_order)]

    customer_orders = {item for item in order if item in products}
    
    return customer_orders 
````

````python
def total_price(customer_orders):
    price = {product : float(input(f"Enter the price of {products}: ")) 
             for product in customer_orders}

    total_price = sum(price.values())
    print(f"Total Price: {total_price}")
    return total_price
````

````python
def update_inventory(customer_orders, inventory):
    update_inventory = {product : (quantity -1 if product in customer_orders else quantity)
                        for product, quantity in inventory.items()}
    
    update_inventory = {product : quantity for product, quantity in inventory.items() if quantity > 0}
    return update_inventory
````

````python
def calculate_order_statistics(customer_orders, products): 
    total_ordered = len(customer_orders)
    percentage = (total_ordered / len(products)) * 100
    return total_ordered, percentage
````

````python
def print_order_statistics(order_statistics):
    total_ordered, percentage = order_statistics
    print("Order Statics")
    print(f"Total unique products ordered: {total_ordered}")
    print(f"Percentage ordered: {percentage}%")  
````

````python
def print_updated_inventory(inventory):
    print("Updated inventory")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")
````

````python
inventory = initialize_inventory(products)
customer_orders = get_customer_orders()
order_stats = calculate_order_statistics(customer_orders, products)
print_order_statistics(order_stats)

inventory = update_inventory(customer_orders, inventory)
print_updated_inventory(inventory)

total_price(customer_orders)
````
